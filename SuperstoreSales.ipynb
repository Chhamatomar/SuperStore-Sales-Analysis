# Univariate Analysis On Numerical data:-

```python
plt.figure(figsize=(5,4))

sns.histplot(data['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()

sns.boxplot(data['Sales'])
plt.title("Sales Outliers Check")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.histplot(data['Profit'], bins=30, kde=True)
plt.title("Profit Distribution")
plt.show
```

```python
(data['Profit']<0).mean()*100
```

```python
plt.figure(figsize=(5,4))
sns.histplot(data['Discount'],bins=20)
plt.title("Discount Distribution")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Quantity', data=data)
plt.title("Quantity Distribution")
plt.show()
```

# Bivariate Analysis:-

```python
plt.figure(figsize=(5,4))

sns.barplot(x='Category', y='Sales', data=data)
plt.title("Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(5,4))

sns.barplot(x='Region', y='Profit', data=data)
plt.title('Profit by Region')
plt.show()
```

```python
plt.figure(figsize=(5,4))

sns.scatterplot(x='Discount',y='Profit',data=data,hue='Category')
plt.title('Profit By Discount')
plt.show()

#There is a negative relation between discount and profit ,meaning higher discount will lead to loss.
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='Quantity', y='Sales', data=data)
plt.title("Sales vs Quantity")
plt.xlabel("Quantity Sold")
plt.ylabel("Sales Amount")
plt.show()

#There is a weak positive correlation between sales and quantity, meaning higher quantity does not always lead to proportionally higher sales.
```

#Monthly/Yearly Trends:-

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Month'] = data['Order Date'].dt.month
data['Year'] = data['Order Date'].dt.year

monthly_sales = data.groupby(['Year','Month'])['Sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend')

#The monthly sales trend shows a steady growth throughout the year with peaks in November and December, possibly due to holiday season sales.
```

# Correlation Between Numerical Columns:-

```python
num_cols = ['Sales', 'Profit', 'Discount', 'Quantity']
corr = data[num_cols].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation between Sales, Profit, Discount & Quantity")
plt.show()

#Sales and Profit show a moderate positive correlation, meaning higher sales generally lead to higher profit.

#Discount has a negative correlation with Profit — heavy discounts reduce profitability.

#Quantity has only a weak correlation with Profit, showing that selling more units doesn’t always increase profit.
```
